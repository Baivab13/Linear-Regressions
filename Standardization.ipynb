```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data =pd.read_csv('Fish.csv')
data
```

```python
x= data[['Length1','Length2','Length3','Height','Width']]
y= data['Weight']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x) 
#Calculate the mean and Standard Deviation for the dataset
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

```python
reg = LinearRegression()
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Predicting with Standardized data
## Standardize the data to be predicted then run the regression

```python
new_data = pd.DataFrame(data=[[20,21,22,10,3],[30,32,31,11,4]],columns=['Length1','Length2','Length3','Height','Width'])
new_data
```

```python
reg.predict(new_data)
```
